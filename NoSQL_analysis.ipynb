### Setup

```python
# import dependencies
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# create an instance of MongoClient
mongo = MongoClient(port=27017)

# create db variable
db = mongo['uk_food']

# create collection variable
establishments = db['establishments']
```

```python
# review a document in the collection
pprint(establishments.find_one())
```

### 1) Which establishments have a hygiene score equal to 20?

```python
# find number of establishments
query = {'scores.Hygiene': 20}
establishments.count_documents(query)
```

```python
# pprint first result
pprint(establishments.find_one(query))
```

```python
# get all matching results
bad_hygiene_places = list(establishments.find(query))

# convert results into a pandas dataframe
bad_hygiene_df = pd.DataFrame(bad_hygiene_places)

# display df
bad_hygiene_df.head(10)
```

###  2) Which establishments in London have a RatingValue greater than or equal to 4?

```python
# find number of establishments
query = {
    'LocalAuthorityName': {
        '$regex': 'London',
        '$options': 'i' #case insensitive
    },
    'RatingValue': {'$gte': 4}
}

establishments.count_documents(query)
```

```python
# convert matching results into a pandas dataframe
good_ratings_df = pd.DataFrame(list(establishments.find(query)))

# display df
good_ratings_df.head(10)
```

### 3) What are the top 5 establishments with a RatingValue of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# find latitude and longitude for Penang Flavours business
new_entry = establishments.find_one({'BusinessName': 'Penang Flavours'})

# capture latitude and longitude
new_entry_lat = new_entry['geocode']['latitude']
new_entry_lng = new_entry['geocode']['longitude']

# display lat and lng
print(f"Latitude: {new_entry_lat}; Longitude: {new_entry_lng}")
```

```python
# find establishments within 0.01 degree of new entry and RatingValue = 5
query = {
    'geocode.latitude': {'$gte': new_entry_lat - 0.01, '$lte': new_entry_lat + 0.01},
    'geocode.longitude': {'$gte': new_entry_lng - 0.01, '$lte': new_entry_lat + 0.01},
    'RatingValue': 5,
    'scores.Hygiene': {'$ne': None}  #exclude null values
}

# sort by lowest hygiene (better hygiene score) and output top 5
sort = [('scores.Hygiene', 1)]
pprint(list(establishments.find(query).sort(sort).limit(5)))
```

```python
# save results to df
top_nearby_entries_df = pd.DataFrame(establishments.find(query).sort(sort).limit(5))
top_nearby_entries_df
```
